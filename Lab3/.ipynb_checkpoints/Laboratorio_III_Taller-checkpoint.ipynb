## 3. Preparación de los datos

Se realiza la separación de los datos de training y de test, en una proporción 80-20 utilizando las imagenes almacenadas en el folder "COVID-19_Radiography_Dataset", el cual contiene 4 folders (que serán las clases) llamados:

```python
image_size=(100,100)
```

```python
batch_size = 8

#Training 70%
train = tf.keras.preprocessing.image_dataset_from_directory(
    "COVID-19_Radiography_Dataset",
    labels='inferred',
    validation_split=0.3,
    subset="training",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
    color_mode="grayscale"
)

val = tf.keras.preprocessing.image_dataset_from_directory(
    "COVID-19_Radiography_Dataset",
    labels='inferred',
    validation_split=0.3,
    subset="validation",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
    color_mode="grayscale"
)

#Validación 20% y Test 10%
val_batches = tf.data.experimental.cardinality(val)
test = val.take(val_batches // 3)
val = val.skip(val_batches // 3)

print('Batches para entrenamiento: %d' % tf.data.experimental.cardinality(train))
print('Batches para validación: %d' % tf.data.experimental.cardinality(val))
print('Batches para test: %d' % tf.data.experimental.cardinality(test))
```

```python
#Estimación de pesos de las clases en el dataset de entrenamiento
labels=np.concatenate([i for x, i in train], axis=0)

class_weights = class_weight.compute_class_weight(class_weight='balanced',classes=np.unique(labels),y=labels)

c_w={}
for i,j in zip(range(4),class_weights):
    c_w[i]=j
    
c_w
```

## 4. Entrenamiento del modelo

#### 4.1 Arquitectura

Para el desarrollo del laboratorio se utilizará una arquitectura Multilayer Perceptron estructurada de la siguiente manera:



```python
h_params={"n_layer1":300,"n_layer2":100,"n_layer4":4,"dropout_1":0.2,"dropout_2":0.2,"dropout_3":0.2}

#Inicialización y construcción de modelo base
model_mlp = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[100, 100]),
    keras.layers.Dropout(h_params["dropout_1"], input_shape=(2,)),
    keras.layers.Dense(h_params["n_layer1"], activation="relu"),
    keras.layers.Dropout(h_params["dropout_2"], input_shape=(2,)),
    keras.layers.Dense(h_params["n_layer2"], activation="relu"),
    keras.layers.Dropout(h_params["dropout_3"], input_shape=(2,)),
    keras.layers.Dense(h_params["n_layer4"], activation="softmax")
])

#Compilamos el modelo usando categorical crossentropy porque es un modelo multietiqueta
model_mlp.compile(loss="sparse_categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy"]) 
```

```python
model_mlp.summary()
```

#### 4.2 Control de complejidad y tiempo de procesamiento
Usaremos 'Early Stopping' para controlar la compljidad y tiempo de procesamiento. Para este caso usaremos como métrica de monitoreo el 'Loss' con paciencia de 3 épocas para verificar que el loss disminuya.

```python
early_stopping = EarlyStopping(monitor='accuracy', patience=5, verbose=1)
callbacks = [early_stopping]
```

#### 4.3 Fit de modelo base
Corremos el modelo base por 30 epochs con un tamaño de batch de 200 y teniendo en cuenta las diferencias de peso de cada una de las clases

```python
history_mlp = model_mlp.fit(train, epochs=30,batch_size = 200,validation_data=val,class_weight=c_w,callbacks=callbacks)
```

```python
#Función para generar el reporte del modelo

def get_train_report(history,model,n_epochs):
    epochs = n_epochs
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']

    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(epochs)

    plt.figure(figsize=(8, 8))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()
    from sklearn.metrics import classification_report, confusion_matrix
    y_pred = model.predict(val)
    predicted_categories = tf.argmax(y_pred, axis=1)
    true_categories = tf.concat([y for x, y in val], axis=0)

    import seaborn as sns
    cm = confusion_matrix(true_categories, predicted_categories)
    fig = plt.figure(figsize = (8,8))
    ax1 = fig.add_subplot(1,1,1)
    sns.set(font_scale=1.4) #for label size
    sns.heatmap(cm, annot=True, annot_kws={"size": 12},
         cbar = False, cmap='Purples',xticklabels=val_class_names,yticklabels=val_class_names);
    ax1.set_ylabel('True Values',fontsize=14)
    ax1.set_xlabel('Predicted Values',fontsize=14)
    plt.show()
```

```python
get_train_report(history_mlp,model_mlp,10)
```

#### 4.4 Búsqueda de hiperparámetros

Definimos un modelo con una capa densa entrada y dropout, una capa densa de salida, y de 1 a 100 capas ocultas con capas de dropout. El modelo será entrenado haciendo el tuning de parámetros con HyperBand dado su rendimiento en tiempo y resultados similares a los obtenidos con RandomSearch (no incluídos en el reporte).

El criterio de optimización del modelo será la exactitud sobre el conjunto de validación, iterando hasta 100 epochs. Se incluye early stopping en el tuning de los modelos para eliminar aquellos que después de cierto número de epochs no mejoren.

```python
def build_model(hp):
    model = Sequential(name='CovidNet')
    model.add(layers.Flatten(input_shape=[100, 100]))
    model.add(Dense(
                units=hp.Int('units', min_value=32, max_value=2048, step=8),
                activation=hp.Choice('dense_activation', values=['relu', 'sigmoid', 'tanh'],default='relu'),
                name='Capa_Entrada'))
    model.add(Dropout(hp.Choice('dropout',[0.1,0.2, 0.3]),name='Dropout'))
    for i in range(hp.Int("num_layers", min_value=1, max_value=100, step=2)):
        model.add(Dense(
                units=hp.Int("units_{0}".format(i), min_value=32, max_value=2048, step=8),
                activation=hp.Choice('dense_activation_{0}'.format(i), values=['relu', 'sigmoid', 'tanh'],default='relu'),
                name= 'Capa_Oculta_{0}'.format(i)))
        model.add(Dropout(hp.Choice('dropout_{0}'.format(i),[0.1,0.2, 0.3]),name='Dropout_{0}'.format(i))) 
        
    model.add(Dense(4, activation='softmax',name= 'Capa_Salida'))
    
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Choice("learning_rate", values=[1e-2, 1e-3])),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model
```

```python
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=100,
    seed=1234,
    directory='tuning',
    project_name='tuningHyper')
tuner.search_space_summary()
```

```python
tuner.search(train, epochs=50,batch_size = 200,validation_data=val,class_weight=c_w,callbacks=[tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=3, verbose=1)])
```

```python
model_tuned = tuner.get_best_models()[0]
best_model = model_tuned
```

```python
tuner.results_summary()
```

```python
prediccion_train = best_model.predict(train)
prediccion_val= best_model.predict(val)
```

```python
print("Resultados para conjunto de entrenamiento")
predicted_categories = tf.argmax(prediccion_train, axis=1)
true_categories = tf.concat([y for x, y in train], axis=0)

cm = confusion_matrix(true_categories, predicted_categories)
fig = plt.figure(figsize = (8,8))
ax1 = fig.add_subplot(1,1,1)
sns.set(font_scale=1.4) #for label size
sns.heatmap(cm, annot=True, annot_kws={"size": 12},
         cbar = False, cmap='Purples',xticklabels=class_names,yticklabels=class_names);
ax1.set_ylabel('True Values',fontsize=14)
ax1.set_xlabel('Predicted Values',fontsize=14)
plt.show()

print(classification_report(predicted_categories, true_categories, target_names=class_names))

print("Resultados para conjunto de validación")
predicted_categories = tf.argmax(prediccion_val, axis=1)
true_categories = tf.concat([y for x, y in val], axis=0)

cm = confusion_matrix(true_categories, predicted_categories)
fig = plt.figure(figsize = (8,8))
ax1 = fig.add_subplot(1,1,1)
sns.set(font_scale=1.4) #for label size
sns.heatmap(cm, annot=True, annot_kws={"size": 12},
         cbar = False, cmap='Purples',xticklabels=class_names,yticklabels=class_names);
ax1.set_ylabel('True Values',fontsize=14)
ax1.set_xlabel('Predicted Values',fontsize=14)
plt.show()

print(classification_report(predicted_categories, true_categories, target_names=class_names))
```

Después del tuning de parámetros, el mejor modelo obtenido no presenta mejoría considerable respecto al modelo base y tiende a clasificar todo en una misma categoría. El mejor modelo obtenido presenta una exactitud sobre el test de entrenamiento y sobre el test de evaluación muy bajos, por lo que no se considera un buen modelo. 

Para obtener un mejor modelo se deben incluir capas convolucionales a la red y algunos pasos de transformación de datos que permitan extraer mejores características o generar datos para balancear las clases antes de entrenar el modelo. Para este entrenamiento se usó la función de class_weight que viene con el modelo como método de balanceo de datos. 

En este caso, esto se fundamenta en las razones por las cuales una red neuronal convolucional generalmente provee un mejor rendimiento como solución al problema de clasificación de imágenes a comparación de un modelo MLP como el planteado en este trabajo. Comenzando por la manera en que las imágenes son procesadas como entradas existe un efecto muy relevante para el resultado final. En el caso de los modelos MLP la cantidad de pesos y parámetros aumenta en una gran medida a medida que la red se hace más profunda, además que las capas se encuentran conectadas por completo, lo cual hace que la red sea difícil de optimizar y computar (Lim, 2017). Este problema puede causar problemas de 'overfitting'. Por otro lado, las redes MLP toman en cuenta la posición de los patrones en la imagen, lo cual, en este caso es impreciso, debido a que las neumonías infecciosas, incluyendo COVID-19, suelen ubicarse en distintos lugares de los lóbulos pulmonares, que, si bien no sucede de forma aleatoria y se relaciona también con las características fisiológicas del paciente y fisiopatológicas del desarrollo de la efermedad, no suelen ser constantes, por lo que este factor pudo ocasionar muchos errores en el proceso de entrenamiento de la red. A comparación, en el caso de las redes convolucionales, su arquitectura está basada de manera que puede simular el procesamiento de imágenes de la corteza visual temprana humana aprovechando las correlaciones espaciales locales en las imágenes (Mnih, 2015). En el caso de estas imágenes, una ventaja de este método es el hecho que las zonas de infiltración pulmonar con procesos inflamatorios causados por la neumonía por COVID-19 presentan patrones muy específicos, por ejemplo leves engrosamientos de vasos sanguíneos, relevantes para la clasificación. Además, este modelo de red permite la consistencia independiente de la traslación de los patrones a través de la imagen, a diferencia de las redes MLP.

Por último, un proceso de 'Data augmentation' pudo haber sido utilizado para disminuir el efecto de 'overfitting' como se reporta en el trabajo de detección de imágenes RX de COVID-19 por Mnih et .al (2015). En este, ejecutaron tres formas de aumentación de datos utilizadas en datasets con tamaños pequeños, como lo son rotaciones hortizontales de las imágenes, rotaciones por ángulos pequeños aleatoriamente e inyección de ruido aleatorio de acuerdo a una distribución estadística para cada pixel de la imagen. Este procedimiento es bastante útil sobretodo en casos donde obtener muestras grandes es difícil de acuerdo con limitaciones de presupuesto, recolección u otros problemas, lo cual sucede frecuentemente en la obtención de imágenes diagnósticas.
